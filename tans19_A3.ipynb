### <font color='red'> A love affair with a nonlinear twist.   </font>

This assignment deals with *love dynamics*, a subject within the field of social psychology. We will consider a simple case, that of a romantic relationship that involves only two individuals. 

Love stories are dynamic processes that start from zero (two persons are completely indifferent to each other when they first meet), develop (more or less quickly) and end up into a sort of regime. Real-life observations tell us that frequently transients develop quite regularly and asymptotic regimes are stationary and associated to positive romantic relationships. But observations also point out the existence of multiple attractors. See [J.C.Sprott, _Dynamical Models of Love_, Nonlinear Dynamics, Psychology, and Life Sciences 8, 303-313 (2004).](http://sprott.physics.wisc.edu/pubs/paper277.htm)

The formal description of love dynamics in a couple requires two state variables, $x$ and $y$, representing the love (or contempt if negative) of each individual for their partner. 

#### A linear love affair 

The simplest model is linear (first introduced by Strogatz (1994))

\begin{align}
\frac{dx}{dt} & = a_xx + b_xy\\
\frac{dy}{dt} & = b_yx + a_yy
\end{align}

where $a_x, a_y, b_x, b_y$ are parameters describing the *romantic style* of each individual: 
- $a_x$ and $a_y$ describe the extent to which each individual is encouraged by their own feelings 
- $b_x$ and $b_y$ describe the extent to which each individual is encouraged by their partner's feelings

Note that $a_x, a_y, b_x, b_y$ can take negative, zero, or positive values.

In some literature, the terms "secure" and "synergic" are used for individuals with $a<0$ and $b>0$, respectively. For instance a *secure individual* ($a<0$) is an individual who retreats from their own feelings and a *non-synergic individual* ($b<0$) is an individual who hates to be loved and loves to be hated. 

The linear system above has an equilibrium point at $(0,0)$ regardless of the choice of parameters $a_x, a_y, b_x, b_y$, and the asymptotic stability of this equilibrium point has been discussed in class in Lecture 10. 

**For instance,** we have seen in class that $(0,0)$ is an "unstable spiral focus" iff $Re(\lambda_d)>0$ and $Im(\lambda_d)\neq0$. That is, iff ${\rm Tr}(A)>0$ and $4\det(A)>{\rm Tr}(A)^2$ (the yellow region in Figure 7.5). #### A nonlinear love affair

Consider the following nonlinear model, based on the assumption that individuals have an adverse reaction if their partner's feelings are excessive. 

\begin{align}
\frac{dx}{dt} & = a_xx + b_xy(1-y^2)\\
\frac{dy}{dt} & = b_yx(1-x^2) + a_yy
\end{align}

\begin{align}
x = 0, y = 0\\
\end{align}

Therefore the point $(0,0)$ is the equilibrium point.\\
